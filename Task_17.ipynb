```python

# Check for duplicate rows
duplicates = df.duplicated().sum()

# Remove duplicate rows
df = df.drop_duplicates()

# Print the new shape of the dataset
print(df.shape)
```

```python
# Count rows with Cholesterol = 0
chol_zero_count = ("Cholesterol = 0:", (df.Cholesterol == 0).sum())

# Count rows with RestingBP = 0
bp_zero_count = ("RestingBP = 0:", (df.RestingBP == 0).sum())

print(chol_zero_count)
print(bp_zero_count)
```

```python
print(df[['Cholesterol', 'RestingBP']].describe())

chol = df[df['Cholesterol'] !=0]
chol_mean = chol['Cholesterol'].mean()
df['Cholesterol'] = df['Cholesterol'].replace(0,chol_mean)

bp = df[df['RestingBP'] !=0]
bp_mean = chol['RestingBP'].mean()
df['RestingBP'] = df['RestingBP'].replace(0,bp_mean)

df['Cholesterol'] = df['Cholesterol'].round(2)
df['RestingBP'] = df['RestingBP'].round(2)

print(df[['Cholesterol', 'RestingBP']].describe())
```

```python
# Before cleaning
print("Before cleaning:")
print(df[['Cholesterol', 'RestingBP']].describe())

# Replace Cholesterol = 0 with mean
chol = df[df['Cholesterol'] != 0]
chol_mean = chol['Cholesterol'].mean()
df['Cholesterol'] = df['Cholesterol'].replace(0, chol_mean)

# Replace RestingBP = 0 with mean
bp = df[df['RestingBP'] != 0]
bp_mean = bp['RestingBP'].mean()
df['RestingBP'] = df['RestingBP'].replace(0, bp_mean)

# Round both columns to 2 decimal places
df['Cholesterol'] = df['Cholesterol'].round(2)
df['RestingBP'] = df['RestingBP'].round(2)

# After cleaning
print("\nAfter cleaning:")
print(df[['Cholesterol', 'RestingBP']].describe())
```

```python
# Create a 2x2 layout
plt.figure(figsize=(20,16))

# Age
plt.subplot(3,2,1)
sns.histplot(df['Age'], kde=True)
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")


# ChestPainType
plt.subplot(3,2,2)
sns.histplot(df['ChestPainType'], kde=True)
plt.title("Distribution of ChestPainType")
plt.xlabel("ChestPainType")
plt.ylabel("Frequency")

# RestingBP
plt.subplot(3,2,3)
sns.histplot(df['RestingBP'], kde=True)
plt.title("Distribution of RestingBP")
plt.xlabel("RestingBP")
plt.ylabel("Frequency")

# MaxHR
plt.subplot(3,2,4)
sns.histplot(df['MaxHR'], kde=True)
plt.title("Distribution of MaxHR")
plt.xlabel("MaxHR")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```
