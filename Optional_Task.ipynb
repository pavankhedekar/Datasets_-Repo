```python
# Age
sns.histplot(df['AGE'], kde=True)
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

```

```python
plt.hist(df['AGE'])
```

```python
# Smoker
plt.subplot(2,2,1)
sns.countplot(x='SMOKING', data=df)
plt.title("Count of Smoker")
plt.xlabel("Smoker")
plt.ylabel("Count")
plt.show()

# ANXIETY
plt.subplot(2,2,2)
sns.countplot(x='ANXIETY', data=df)
plt.title("Count of ANXIETY")
plt.xlabel("ANXIETY")
plt.ylabel("Count")
plt.show()
```

```python
# Identify numerical columns.
numerical_cols = df.select_dtypes(include = 'number').columns
numerical_cols
```

```python
# Heatmap
corr = df[numerical_cols].corr()
sns.heatmap(corr)
plt.title("Correlation Heatmap")
plt.show()

```

```python
# Boxplots or use hue parameter
sns.boxplot(x='SMOKING', y='ANXIETY', hue='AGE',data = df)
```
